## Array Attributes

```python
import numpy as np
a1=np.arange(10)
a2=np.arange(12).reshape(4,3)
a3=np.arange(8).reshape(2,2,2)

print(f"a1:\n{a1}")
print()
print(f"a2:\n{a2}")
print()
print(f"a3:\n{a3}")
print()
```

```python
# ndim
print(a1.ndim) # of dimentions
print(a2.ndim)
print(a3.ndim)
```

```python
#shape
print(a1.shape) #tells about te shape of the array
print(a2.shape)
print(a3.shape)
```

```python
# size
print(a1.size) # tells about the size of the array
print(a2.size)
print(a3.size)
```

```python
# itemsize
print(a1.itemsize) # tells that how much memory size is talking every item
print(a2.itemsize) # all are int 
print(a3.itemsize)
```

```python
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
```

## Array Operations

```python
var=arry **2 # takes the square of the entire array
print(var)
```

```python
# relational Operators
print(var > 5)
```

#### Array Functions

```python
A1=np.random.random((4,4)) # 4 rows and 4 columns
round=np.round(A1 * 100)
print(round)
```

```python
# max/min/sum/prod
max=np.max(round)
min=np.min(round)
print(max)
print(min)
print(np.sum(round)) # gives the sum of the all items
print(np.prod(round)) # gives the product of the all elements
```

```python
#  for axis row=1 and col=0 
print(f"Original Matrix:\n {round}\n")

column_wise_max=np.max(round, axis=0)
row_wise_max=np.max(round, axis=1)
print(f"Column wise max:\n{column_wise_max}\n")
print(f"Row wise max:\n {row_wise_max}")
```

#### Mean/Median/Std/Var

```python
import numpy as np
mean=np.mean(round,axis=1) # row wise mean
median=np.median(round,axis=1)
standard_deviation=np.std(round,axis=0) # column wise std
varience=np.var(round,axis=0) # column wise varience is taken 
print(mean)
print(median)
print(standard_deviation)
print(varience)
```

```python
# trignomatrix fumction
np.sin(round) # similar can be done others
```

```python
# dot product of two matrix
mat1=np.arange(1,11).reshape(5,2)
mat2=np.arange(20,30).reshape(2,5)
# now dot product 
product=np.dot(mat1,mat2)
product
```

```python
# log and exponents
log=np.log(mat1)
exp=np.exp(mat1)
print(f"{log}\n")
print(exp)
```

```python
# round, floor and ceil

round=np.round(np.random.random((2,3))*100)
floor=np.floor(np.random.random((2,3))*100)
ceil=np.ceil(np.random.random((2,3))*100)

print(f"Round:\n {round}\n")
print(f"Floor:\n {floor}\n")
print(f"Ceil:\n {ceil}\n")
```
